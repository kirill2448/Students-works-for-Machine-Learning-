```python
import numpy as np
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Nadam
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python



F = list(range(-200,200))
C = [(f - 32) * 5 / 9 for f in F]

model = Sequential([Dense(1, activation='linear', input_shape=[1])])
model.compile(loss='mean_squared_error', optimizer=Nadam(lr=0.001), metrics=['mean_absolute_error'])


history=model.fit(F, C, batch_size=16, epochs=500)

print(model.predict(np.array([300])))
print(model.layers[0].get_weights())
plt.plot(history.history['loss'])
plt.grid(True)
plt.show()
```
