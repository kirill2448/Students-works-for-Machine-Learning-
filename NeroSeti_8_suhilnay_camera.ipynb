```python
from sklearn.model_selection import train_test_split
X_train1 ,X_test1,y_train1,y_test1=train_test_split(
data[:,:-1],
data[:,-1],
test_size=0.20    
)
```

```python
model = Sequential([
                   Dense(11,activation='linear', input_shape=[10]),
                   Dense(9, activation='sigmoid', input_shape=[10]),
                   Dense(5, activation='linear', input_shape=[10]),
                   Dense(3, activation='linear',input_shape=[10]),
                   Dense(1, activation='linear', input_shape=[10])
                         ])
model.compile(loss='mean_squared_error', optimizer=Nadam(lr=0.0001), metrics=['mean_absolute_error'])

X_train1 = np.asarray(X_train1).astype(np.float32)

y_train1 = np.asarray(y_train1).astype(np.float32)

X_test1 = np.asarray(X_test1).astype(np.float32)

history=model.fit(X_train1, y_train1, batch_size=25, epochs=1000)

res = model.predict(X_test1, verbose = 0)
res = np.array(res)
res = res.reshape(np.max(res.shape), )
detres = res * scaler.scale_[-1] + scaler.mean_[-1]
detsource = y_test1 * scaler.scale_[-1] + scaler.mean_[-1]

print(detres)
print(model.layers[0].get_weights())
plt.plot(history.history['loss'])
plt.grid(True)
plt.show()
```
